```python
archivo = "/content/drive/MyDrive/CICLO 2022 - 1/Sisitema de Informaciòn Geografica/Examen Final/LIMITE_DISTRITAL_2020_INEI_geogpsperu_juansuyo_931381206.shp"
```

```python
mapa = gpd.read_file(archivo)
mapa.head()
```

```python
DEP = input("Departamento:")
nombre = DEP
```

```python
DEP = mapa.loc[mapa["NOMBDEP"] == DEP]
DEP
```

```python
colum_sup = list(DEP.columns)
print(colum_sup)
```

```python
colum_sup.remove("NOMBDEP")
colum_sup.remove("NOMBDIST")
colum_sup.remove("geometry")
```

```python
DEP = DEP.drop(labels = colum_sup, axis = 1)
DEP
```

```python
DEP["centroid"] = DEP["geometry"].centroid
DEP
```

```python
fig, ax=plt.subplots(1, figsize=(22,8))
legend_config={
                "title":"distritos",
                "loc":"upper left",
                "bbox_to_anchor":(1,1.03),
                "ncol": 3}
DEP.plot(column="NOMBDIST",
                          cmap="OrRd",
                          k=20,
                         legend=True,
                         legend_kwds=legend_config, ax=ax)

props=dict(boxstyle="round", facecolor="linen",alpha=1)

for point in DEP.iterrows():
  ax.text(point[1]["centroid"].x,
          point[1]["centroid"].y,
          point[1]["NOMBDIST"],
          horizontalalignment="center",
          fontsize=5,
          bbox=props)
ax.axis("on")

plt.title("Mapa distrital del departamento de arequipa",fontdict={"fontsize":"15"})
plt.xlabel("Este")
plt.ylabel("Norte")

#para guardar el mapa
plt.savefig("map_categorizado_arequipa.png",dpi=500)

plt.show()
```

##DISOLVE

```python
mapa = gpd.read_file(archivo)
mapa
```

```python
departamentos = mapa.dissolve(by="NOMBDEP")
departamentos
```

```python
departamentos.plot()
plt.show()
```

```python
suda = "/content/drive/MyDrive/CICLO 2022 - 1/Sisitema de Informaciòn Geografica/Examen Final/Copia de Sudamérica.shp"
```

```python
sudamerica = gpd.read_file(suda)
sudamerica
```

```python
sudamerica.plot()
plt.show()
```

```python
#SEABORN
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1, rc={"lines.linewidth": 2})

fig, ax=plt.subplots(1, figsize=(22,8))
#se coloca de color gris los shapefiles de departamentos
departamentos.plot(ax = ax,color="gray")
sudamerica.plot(ax = ax, color="gray")
legend_config={
                "title":"DISTRITOS",
                "loc":"upper left",
                "bbox_to_anchor":(1,1.03),
                "ncol": 3}
DEP.plot(column="NOMBDIST",
                          cmap="gist_earth",
                          k=20,
                         legend=True,
                         legend_kwds=legend_config, ax=ax)

props=dict(boxstyle="round", facecolor="linen",alpha=1)

#BORDES DE PIURA
DEP.geometry.boundary.plot(ax = ax, edgecolor = "black", linewidth = .75)

#BORDES DE LOS DEPARTAMENTOS
departamentos.geometry.boundary.plot(ax = ax, edgecolor = "black", linewidth = .75)

#BORDES DE SUDAMERICA
sudamerica.geometry.boundary.plot(ax = ax, edgecolor = "black", linewidth = .1000)

#LIMITES DE LA GRAFICA
ax.set_ylim(-6.5, -4)
ax.set_xlim(-81.5, -79)

props=dict(boxstyle="round", facecolor="linen",alpha=1)
for point in DEP.iterrows():
  ax.text(point[1]["centroid"].x,
          point[1]["centroid"].y,
          point[1]["NOMBDIST"],
          horizontalalignment="center",
          fontsize=5,
          bbox=props)
ax.axis("on")

plt.title("MAPA DISTRITAL DE PIURA",fontdict={"fontsize":"15"})
plt.xlabel("Este")
plt.ylabel("Norte")

#GUARDAR
plt.savefig("MAPA_PIURA.png",dpi=500)

plt.show()
```
